```python
def classify_article(article):
    doc = nlp(article['title'])
    for ent in doc.ents:
        if ent.label_ == "POSITIVE":
            return 'POSITIVE'
    return 'NEGATIVE'
```

```python
from collections import OrderedDict

def process_articles(news, in_href):
    global articles
    unique_articles = OrderedDict()

    for article in news:
        sentiment = classify_article(article)
        article_link = article['link']
        if sentiment == 'POSITIVE':
            unique_articles[article_link] = None

    for link in in_href:
        unique_articles[link] = None

    articles = list(unique_articles.keys())


process_articles(news, in_href)

print(articles)
```

```python
import requests
from bs4 import BeautifulSoup
import re

# Initialize an empty list for tickers
flat_list = []

# Function to get active stocks
def get_active_stocks():
    url = 'https://www.allpennystocks.com/hotpennystocks/nsd.aspx'
    response = requests.get(url)
    html_content = response.text

    soup = BeautifulSoup(html_content, 'html.parser')
    articles = soup.find_all('tr')

    links_list = []
    ticker_set = set()

    for article in articles:
        cells = article.find_all("td")
        for cell in cells:
            link = cell.find('a', href=True)
            if link:
                links_list.append(link['href'])

    for link in links_list:
        upper_letters = re.findall(r'[A-Z]+', link)
        ticker_set.update(upper_letters)

    return list(ticker_set)

flat_list.extend(get_active_stocks())

print(flat_list)
```

```python
import requests
from bs4 import BeautifulSoup
import re
from concurrent.futures import ThreadPoolExecutor

articles


Tickers=[]
with ThreadPoolExecutor(max_workers=4) as executor:
    futures = [executor.submit(requests.get, article) for article in articles]

    for future in futures:
        response = future.result()
        print(response.status_code)

        html_content = response.content
        soup = BeautifulSoup(html_content, 'html.parser')

        bold_tags = soup.find_all(['b', 'strong'])
        text_content = ' '.join(tag.get_text(strip=True) for tag in bold_tags)

        ticker_pattern = re.compile(r'\b[A-Z]{3,8}\b')
        tickers = ticker_pattern.findall(text_content)
        unique_list = list(set(tickers))
        print(f"Extracted tickers: {unique_list}")
        Tickers.append(unique_list)



```

```python
flat_list = [item for sublist in Tickers for item in sublist]
flat_list
```

```python
combined_list = flat_list + get_active_stocks()
```

```python
import yfinance as yf # Make sure you have yfinance installed: !pip install yfinance
import pandas as pd # Import pandas with the alias 'pd'

under_5_stocks = []
for stock_name in combined_list:
    data = yf.download(stock_name, period='1d')
    # Check if DataFrame is empty and 'Close' column exists
    if not data.empty and 'Close' in data.columns:
        latest_price = data['Close'].iloc[-1] # Use iloc for integer-based indexing
        # Check if latest_price is a Series and if so, get the first value
        if isinstance(latest_price, pd.Series):
            latest_price = latest_price.iloc[0]
        if latest_price <= 10:
            under_5_stocks.append(stock_name)
            print("Stock:", stock_name)
```

```python
existing_tickers = set()

with open('under_5tickers.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        existing_tickers.update(row)


new_tickers = [ticker for ticker in under_5_stocks if ticker not in existing_tickers]

if new_tickers:
    with open('under_5tickers.csv', 'a', newline='') as file:
        writer = csv.writer(file)
        for ticker in new_tickers:
            writer.writerow([ticker])
        print("Tickers added successfully:", new_tickers)
else:
    print("All tickers already exist.")
```
